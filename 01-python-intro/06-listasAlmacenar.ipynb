```python
lista_letras=["b","a","e","d"]
lista_letras_2=["c","z","h"]
for letra in lista_letras_2:
    #print(letra)
    lista_letras.append(letra)

print(lista_letras)
```

```python
### para unir listas
lista_letras=["b","a","e","d"]
lista_letras_2=["c","z","h"]
lista_letras.extend(lista_letras_2)
print(lista_letras)
```

### Metodos de listas
- insert -> inserta un elemento en la posicion o indice dado
- append -> inserta un elemento en la ultima posicion
- clear -> borra el contenido de mi lista
- pop -> elimina el ultimo item de mi lista
- extend -> agrega los elemntos de una lista a otra lista
- sort -> ordenad los elemntos de una lista
- count -> cuenta el valor que se le de en un lista
- len -> sirve para saber cuantos elementos tiene mi lista
- index -> para ver que posicion ocupa el elemento en una lista
