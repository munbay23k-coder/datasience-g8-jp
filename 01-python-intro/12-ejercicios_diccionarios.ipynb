```python
# ejercicio 1
inventario_norte = {"laptop": 4, "mouse": 10, "monitor": 3}
inventario_sur = {"laptop": 2, "teclado": 8, "monitor": 5}
inventario_centro = {"mouse": 6, "teclado": 4, "webcam": 7}

```

```python
inventario_total=inventario_norte.copy()


for llave,valor in inventario_sur.items():
    if llave not in inventario_total:
        inventario_total[llave]=0
        
    inventario_total[llave]+= valor

for llave,valor in inventario_centro.items():
    if llave not in inventario_total:
        inventario_total[llave]=0
        
    inventario_total[llave]+= valor

print(inventario_total)
```

```python
llaves_ordenadas=sorted(inventario_total.keys())
inventario_total_con_llaves_ordenadas={}
for llave in llaves_ordenadas:
    inventario_total_con_llaves_ordenadas[llave]=inventario_total[llave]

print(inventario_total)
print(inventario_total_con_llaves_ordenadas)
```

```python
# Ejercicio 2
calificaciones = {
 "Ana": [17, 18, 16, 19],
 "Luis": [15, 16, 17, 16],
 "Carla": [18, 17, 19, 18],
 "Pedro": [16, 16, 16, 16],
 "Lucía": [15, 17, 16, 16]
}
```

```python
calificaciones_corregidas=calificaciones.copy()
estadisticas={}
for alumno,notas in calificaciones_corregidas.items():
    estadisticas[alumno]={
        "minima_nota":min(notas),
        "maxima_nota":max(notas),
        "promedio":sum(notas)/len(notas)

    }
print(estadisticas)
```

```python
for estudiante, stats in estadisticas.items():
    print(estudiante)
    print(stats)
    print('='*20)
```

```python
ranking=[]
for nombre,stats in estadisticas.items():
    ranking.append((nombre,stats["promedio"]))## esto es para hacer tuplas los parentesis

ranking=sorted(ranking, key=lambda x: (x[1], x[0]), reverse=True)### x[1]  1 significa que ordene segun elemento 2 que es promedio
print(ranking)
```

```python
ranking[2:] = sorted(ranking[2:],key=lambda x: x[0]) ### x[0] 0  significa que ordene segun elemento 1 que es nombres
```

```python
text_sin_palabras=[]
igual=0
for text in texto_nuevo:
    igual=0
    for palabra in palabras_vacias:
        if text == palabra :
            igual=1
    if igual==0:
        text_sin_palabras.append(text)

print(text_sin_palabras)
```

```python
mi_palabras={}
for texto in text_sin_palabras:
    cantidad=0
    for texto2 in text_sin_palabras:
        if texto == texto2:
            cantidad+=1
    mi_palabras.update({texto:cantidad})
```

```python
ordenado=dict(sorted(mi_palabras.items()))
print(ordenado)
```

```python
### ejercicio 5
empleados = [
 {"nombre": "Ana", "departamento": "Desarrollo", "nivel": "Senior", "salario":
5200},
 {"nombre": "Luis", "departamento": "Diseño", "nivel": "Junior", "salario": 2800},
 {"nombre": "Carla", "departamento": "Desarrollo", "nivel": "Junior", "salario":
3400},
 {"nombre": "Pedro", "departamento": "Desarrollo", "nivel": "Senior", "salario":
5600},
 {"nombre": "Lucía", "departamento": "Diseño", "nivel": "Senior", "salario": 4700}
]
```

```python
resumen={}

for idx,empleado in enumerate(empleados):#### para tener el dato del numero de dato es
    departamento=empleado['departamento']
    nivel= empleado['nivel']
    nombre=empleado['nombre']
    salario=empleado['salario']
    if departamento not in resumen:
        resumen[departamento]={}

    if nivel not in resumen[departamento]:
        resumen[departamento][nivel]={
                "cantidad":0,
                "nombres":[],
                "salarios":[]
            }

    grupo= resumen[departamento][nivel]
    grupo["nombres"].append(nombre)
    grupo["salarios"].append(salario)
    grupo["cantidad"]+=1

for registro ,grupo_xp in resumen.items():
    for experiencia,valor in grupo_xp.items():
        valor["promedio"]= sum(valor["salarios"])/len(valor['salarios'])
```

```python
#ejercicio 6
tienda_a = {"laptop": 3500, "mouse": 80, "teclado": 160, "monitor": 920} 
tienda_b = {"laptop": 3400, "mouse": 90, "teclado": 150, "webcam": 220} 
```

```python
productos=list(tienda_a.keys())
productos.extend(list(tienda_b.keys()))## para agregar los productos que hay en una lista
productos=set(productos)##para que no se repitan los nombres de los productos

comparativo={}

for producto in productos:
    if producto in tienda_a and producto in tienda_b:

        if tienda_a[producto]>tienda_b[producto]:
            mas_barato="tienda_b"
            precio = tienda_b[producto]
        elif tienda_a[producto]==tienda_b[producto]:
            mas_barato="es igual en ambas tiendas"
            precio = tienda_a[producto]
        else:
            mas_barato="tienda_a"
            precio = tienda_a[producto]

        comparativo[producto]={
            "diferencia":abs(tienda_a[producto]-tienda_b[producto]),
            "mas_barato": mas_barato,
            "precio":precio
        }

    if producto in tienda_a and producto not in tienda_b:
        comparativo[producto]={
            "diferencia":"no hay diferencia",
            "mas_barato": "tienda_a",
            "precio":tienda_a[producto]
        }
    if producto not in tienda_a and producto  in tienda_b:
        comparativo[producto]={
            "diferencia":"no hay diferencia",
            "mas_barato": "tienda_b",
            "precio":tienda_b[producto]
        }

for com in comparativo:
    print(com)
    print(comparativo[com])
    print("="*10)
```

```python
## ejercicio 7
candidatos = {"Ana", "Luis", "Carla"} 
votos = ["Ana", "Luis", "ana", "Carla", "Pedro", "Luis", "ANA", "", "Carla"]

mapa ={
    "ana":"Ana",
    "luis":"Luis",
    "carla":"Carla"
}

conteo={}

for voto in votos:
    voto_limpio=voto.lower()
    candidato=mapa.get(voto_limpio, "invalido")## si no esta el nombre o esta vacio devolvera invalido

    if candidato=="invalido":
        conteo.setdefault("votos invalidos",0)
        conteo["votos invalidos"]+=1
        continue

    conteo.setdefault(candidato,0)
    conteo[candidato]+=1

total_votos=len(votos)
for candidato in candidatos:
    conteo[f"{candidato}_pct"]=round(conteo[candidato]/total_votos*100,2)

conteo["votos invalidos pct"]=round(conteo["votos invalidos"]/total_votos*100,2)
```
