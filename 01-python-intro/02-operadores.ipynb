# Funciones u operadores en string

- Longitud de una cadena: `len()`
- Convertir a mayúsculas: `upper()`
- Convertir a minúsculas: `lower()`
- Reemplazar un valor de la cadena: `replace()`
- Eliminar espacios al inicio y al final: `"texto".strip()`
- Dividir una cadena: `"texto1,texto2,texto3".split(",")` → `["texto1", "texto2", "texto3"]`

```python
nombre= "Macell     "
print(len(nombre))
nombre_limpio= nombre.strip()
print(len(nombre_limpio))
```

```python
nombres= "Jhair Perez"
nombres_separados=nombres.split()
print(nombres_separados)

nombres="JhairxxPerez"
nombres_separados=nombres.split("xx")
print(nombres_separados)
```
