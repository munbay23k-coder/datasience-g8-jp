```python
## ejercicio 4
tupla= (18,15,20,11,14,17,19)
promedio=sum(tupla)/len(tupla)
maxima_nota=max(tupla)
minima_nota=min(tupla)
lista_aprobados=[
    nota
    for nota in tupla
        if nota > 10
]
print(f"el promedio es:{promedio}")
print(f"el valor maximo es:{maxima_nota}")
print(f"el valor minimo es:{minima_nota}")
print(f"La cantida de aprobados es:{len(lista_aprobados)}")
```
