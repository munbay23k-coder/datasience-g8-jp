```python
# Encapsular- controlar algunos atributos de mis datos

class Cuenta:
    def __init__(self,saldo:float):### metodo iniciar
        self.__saldo=saldo### se hace atributo privado

    def depositar(self,monto:float):##metodo depositae
        self.__saldo +=monto## se hace atributo privado con(__)
    
    def retirar(self,monto:float):
        if monto> self.__saldo:
            return "Error saldo insuficiente"
        self.__saldo -=monto
    
    def mostrar_saldo(self):## aca se devuelve el valor pero tu no accedes al atributo
        return self.__saldo
    
cuenta=Cuenta(1000)
print(cuenta.mostrar_saldo())
        
```

```python
# Herencia
class Animal:
    def __init__(
        self,
        es_domestico: bool,
        num_patas: int,
        nombre: str                 
    ):
        self.es_domestico=es_domestico
        self.num_patas=num_patas
        self.nombre= nombre
    
    def comer(self):
        print("Estoy comiendo")

    def beber(self):
        print("Estoy bebiendo agua")

    def es_adoptable(self):
        if self.es_domestico:
            return True
        return False

class Duenio:
    def __init__(self,nombre:str,edad:int):
        self.nombre=nombre
        self.edad=edad

    def adoptar(self):
        return "Estoy adoptando"
        
class Perro(Animal):# se hereda los atributos y metodos de animal
    def __init__(
        self,
        es_domestico: bool,
        num_patas: int,
        nombre: str  ,
        nombre_duenio:str,
        edad_duenio:int               
    ):
        super().__init__(es_domestico,num_patas,nombre)#Sirve para llamar al constructor (__init__) de la clase padre
        self.duenio=Duenio(nombre_duenio,edad_duenio) ## es composision esto porque no se puede heredar de duenio a perro
    def hablar(self):
        return "gua gua!"
        
class Gato(Animal):# se hereda los atributos y metodos de animal
    def __init__(
        self,
        es_domestico: bool,
        num_patas: int,
        nombre: str                 
    ):
        super().__init__(es_domestico,num_patas,nombre)
    def hablar(self):
        return "miau miau!"  
    def beber(self):
        print("yo no bebo agua , yo tomo leche")  
```
