```python
## ejercicio 1
numeros = [4, 7, 2, 4, 9, 7, 1, 2, 10, 4]
numeros_distintos=[]
for numero in numeros:
    if numero not in numeros_distintos:
        numeros_distintos.append(numero)
print(numeros_distintos)
```

```python
##ejercicio 2
numeros = [12, 7, 9, 24, 31, 18, 5, 42, 13, 8]
numeros_pares = [
    numero
    for numero in numeros
    if numero % 2 == 0
]

numeros_impares = [
    numero
    for numero in numeros
    if numero % 2 == 1
]
print(numeros_pares)
print(numeros_impares)
```

```python
## ejercicio 16
numeros = [10, 20, 30, 40, 50, 60, 70, 80]
half_list=[]
last_list=[]
longitud_lista=len(numeros)
indice_mitad=longitud_lista/2
print(indice_mitad)
for idx, numero in enumerate(numeros):
    if idx < indice_mitad:
        half_list.append(numero)
    else:
        last_list.append(numero)
print(f"Primera mitad: {half_list}")
print(f"Segunda mitad: {last_list}")
```
