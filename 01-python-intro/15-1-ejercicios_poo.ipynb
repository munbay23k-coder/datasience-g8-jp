```python
## Ejercicio 1
class Persona:
    def __init__(self,nombre:str,edad:int)->None:## sirve para inicializar la clase
        self.nombre=nombre
        self.edad=edad
    def saludar(self): ## crear metodo
        print(f"Hola, soy {self.nombre} y tengo {self.edad} anhos")

jorge = Persona("Jorge",10)## crear objeto jorge
maria= Persona("Maria",20)
luis= Persona("Luis",15)

jorge.saludar()
maria.saludar()
luis.saludar()
```

```python
# Ejercicio 2
class Rectangulo:
    def __init__(self,base:float,altura:float):
        self.base=base
        self.altura=altura
    def area(self)->float:
        return self.base*self.altura
    
    def perimetro(self)->float:
        return 2*(self.base +self.altura)
    
    def __str__(self)->str:
        return (f"Mi rectangulo de base {self.base} y altura {self.altura}")
        
    
mi_rectangulo=Rectangulo(3,5)
print(mi_rectangulo.area())
print(mi_rectangulo.perimetro())
print(mi_rectangulo)### busca el metodo __str__
```
