- UNION o |

```python
#Ejercicio 1
estudiantes_python = {"Ana", "Luis", "María", "Carlos", "Sofía", "Diego"}
estudiantes_javascript = {"María", "Carlos", "Pedro", "Lucía", "Diego", "Valeria"}

print(f"Estudiantes en ambos cursos",estudiantes_python.intersection(estudiantes_javascript))
print("Los estudiantes que solo estan en python",estudiantes_python.difference(estudiantes_javascript))
print("Los estudiantes que solo estan en javascript",estudiantes_javascript.difference(estudiantes_python))
print("Todos los estudiantes son:",estudiantes_python | estudiantes_javascript)
```

```python
# Ejercicio 2
almacen_arequipa = {"laptop", "teclado", "mouse", "monitor", "impresora", "audifonos"}
almacen_lima = {"laptop", "mouse", "monitor", "tablet", "webcam", "parlantes"}
print("Los productos exitentes en ambos almacenes:",almacen_arequipa & almacen_lima)
print("Los productos exclusivos del almacen de arequipa:",almacen_arequipa - almacen_lima)
print("Los productos exclusivos del almacen de lima:", almacen_lima - almacen_arequipa)
print("Todos los productos disponibles sin repetir:",almacen_arequipa | almacen_lima)
print("Si ambos almacenes tienen los mismos productsos", almacen_arequipa == almacen_lima)
```

```python
# Ejercicio 3
permisos_usuario = {"leer", "editar", "comentar", "descargar"}
permisos_requeridos = {"leer", "editar", "eliminar", "publicar"}
permisos_totales = list(permisos_usuario | permisos_requeridos)
print("El usurio tiene todos los permisos:", permisos_requeridos.issuperset(permisos_usuario))
print("Que permisos le fantan al usurio:", permisos_requeridos - permisos_usuario)
print("Que permisos adicionales tiene:", permisos_usuario - permisos_requeridos)
print("La cantidad de permisos son:", len(permisos_totales))
```

```python
# Ejercicio 4
numeros_a = {2, 4, 6, 8, 10, 12, 14}
numeros_b = {4, 6, 9, 10, 13, 14, 16}
numeros_c = {1, 4, 6, 10, 11, 14, 18}

print("Los números presentes en los tres conjuntos",numeros_a & numeros_b & numeros_c)
print("Los números presentes en numeros_a y numeros_b, pero no en numeros_c",numeros_a | numeros_b - numeros_c)
print("Los números que solo pertenecen a numeros_a", numeros_a - numeros_b - numeros_c)
print("Todos los números diferentes",(numeros_a^numeros_b)|(numeros_a^numeros_c)|(numeros_c^numeros_b))
```

```python
# Ejercicio 5
habilidades_candidato_1 = {"Python", "Git", "SQL", "Docker", "React"}
habilidades_candidato_2 = {"Python", "JavaScript", "React", "Node.js", "Git"}
habilidades_requeridas = {"Python", "Git", "SQL", "Docker"}
habili_reque1=habilidades_requeridas - habilidades_candidato_1
habili_reque2=habilidades_requeridas - habilidades_candidato_2
candidato=0
if len(habili_reque2)>len(habili_reque1):
    candidato=1
else: 
    candidato=2
print("Las habilidades que comparten ambos candidatos",habilidades_candidato_1 & habilidades_candidato_2)
print("Las habilidades exclusivas de cada candidato",habilidades_candidato_2 ^ habilidades_candidato_1)
print("Qué habilidades requeridas le faltan a cada candidato 1:",habilidades_requeridas - habilidades_candidato_1)
print("Qué habilidades requeridas le faltan a cada candidato 2:",habilidades_requeridas - habilidades_candidato_2)
print("Qué candidato cumple con más requisitos es el numero:",candidato)
print("Todas las habilidades disponibles sin repetir",habilidades_candidato_1 | habilidades_candidato_2 | habilidades_requeridas)
```
