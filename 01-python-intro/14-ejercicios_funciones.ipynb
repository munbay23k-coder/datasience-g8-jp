```python
## ejercicio 1

def analizar_ventas(ventas : list[float|int])->dict[str:int|float]:
    if not ventas:
        return "no hay ventas"
    
    for venta in ventas:
        if venta < 0:
            return f"Hay ventas negativas: {venta}"
        
    total_vendido=sum(ventas)
    promedio=sum(ventas)/len(ventas)
    venta_mas_alta=max(ventas)
    venta_mas_baja=min(ventas)

    total_superior_promedio=0
    total_inferior_100=0
    for venta in ventas:
        if venta > promedio:
            total_superior_promedio=total_superior_promedio+1
        if venta < 100:
            total_inferior_100=total_inferior_100+1
    
    return{
        "total_vendido":total_vendido,
        "promedio":promedio,
        "venta_mas_alta":venta_mas_alta,
        "venta_mas_baja":venta_mas_baja,
        "total de ventas superior al promedio":total_superior_promedio,
        "total ventas inferior a 100":total_inferior_100
    }

analizar_ventas(ventas)
```

```python
## ejercicio 2

def valiadar_password(password: str) -> tuple[bool, list[str]]:
    errores=[]
    simbolos_requeridos=[ "@", "#" , "$","%"]
    if len (password)<8:
        errores.append(
            "El password tiene menos de 8 caracteres"
        )

    contiene_minuscula =  False
    contiene_mayuscula = False
    contiene_numero = False
    contiene_simbolos_requeridos = False

    for caracter in password:
        if caracter.islower():
            contiene_mayuscula=True
            break

    for caracter in password:
        if caracter.isupper():
            contiene_minuscula=True
            break

    for caracter in password:
        if caracter.isnumeric():
            contiene_mayuscula=True
            break 

    for caracter in password:
        if caracter in simbolos_requeridos:
            contiene_simbolos_requeridos=True
            break 

    if not contiene_minuscula:
        errores.append(
            "El password no contiene letra minuscula"
        )

    if not contiene_mayuscula:
        errores.append(
            "El password no contiene letra mayuscula"
        )
    
    if not contiene_numero:
        errores.append(
            "El password no contiene caracter numerico"
        )
    
    if not contiene_simbolos_requeridos:
        errores.append(
            "El password no contiene los simbolos requeridos"
        )
    
    if len(password) != len(password.strip()):
        errores.append(
            "El password contiene espacios vacios"
        )

    es_valida = False if errores else True

    return(es_valida,errores)

print(valiadar_password("Alvaro  "))
```

```python
def agrupar_por_longuitud(palabras:list[str]) -> dict[int,list[str]]:
    palabras_limpias=set([
        palabra.lower()
        for palabra in palabras
        if isinstance(palabra,str)
    ])
    diccionario={}
    for palabra in palabras_limpias:
        longitud = len(palabra)
        diccionario.setdefault(longitud, []).append(palabra)
        diccionario = dict(sorted(diccionario.items()))
    for llave in diccionario:
        diccionario[llave].sort()    
    return diccionario

    

palabras = [ 
'python', 'Java', 'código', 'web', 
'java', 'datos', 'API', 'lista' 
] 
print(agrupar_por_longuitud(palabras))
```

```python
# Ejercicio 5


def actualizar_inventario(inventario:dict[str,int],producto:str,cantidad:int,operacion:str)->dict[str,str|int]:
    operaciones=("entrada","salida")#tupla para no modificar 
    operacion=operacion.lower()
    if operacion not in operaciones:
        return {
            "mensaje":f"Error, tipo de operacion no soportada: {operacion}"
        }
    if not isinstance(cantidad,int) :
        return {
            "mensaje":f"Error, tipo de dato no adecuada "
        }
    if cantidad <=0 :
        return {
            "mensaje":f"Error, la cantidad es negativa o cero"
        }
    if operacion == "entrada":
        if producto not in inventario:
            inventario[producto]=cantidad
        else:
            inventario[producto]+=cantidad
    if operacion == "salida" and producto in inventario and inventario[producto]>=cantidad:
        inventario[producto]=inventario[producto]-cantidad

    return "Comprobacion exitosa"
    
```

```python
inventario = { 
'Laptop': 5, 
'Mouse': 12, 
'Teclado': 8 
} 
```

```python
resultado=actualizar_inventario(inventario,'Mouse',4,"salida")
print(inventario)
print(resultado)
```

```python
# Ejercicio 6

def calcular_pedido(productos:dict[str:int],cantidades:dict[str:int],descuento:int,delivery:int)->dict[dict[str:int],list[str]]:
    detalles_pedido={}
    no_encontrados=[]
    sub_total=0
    total=0
    for producto,cantidad in cantidades.items():
        if not(descuento>0 and descuento<100):
            return("el descuento no esta en el rango de 0 a 100")
            break
        if not isinstance(delivery,int) or delivery < 0:
            return("el delivery no es entero positivo")
            break
        if not isinstance(cantidad,int) or cantidad < 0:
            return(f"la cantidad de {producto}: {cantidad} no es entera positiva")
            break
        if producto in productos:
            sub_total+=productos[producto]*cantidad
            
        else:
            no_encontrados.append(producto)
        total=delivery+sub_total*(100-descuento)/100
    detalles_finales={
        "subtotal":sub_total,
        "descuento(%)":descuento,
        "delivery":delivery,
        "total":total
    }
    detalles_pedido.setdefault("pedido",detalles_finales)
    detalles_pedido.setdefault("productos no encontrados",no_encontrados)  

    return detalles_pedido
    
```

```python
productos = {'Hamburguesa': 15, 'Pizza': 25, 'Gaseosa': 5} 
cantidades = {'Hamburguesa': 2, 'Gaseosa': 3, 'Helado': 1}
calcular_pedido(productos,cantidades,20,10)
```

```python
#Ejercicio 7

def evaluar_empleados(empleados:dict[str:list[int|float]],minimo:int)->list:
    promedio=0
    lista_empleados=[]
    for empleado,puntajes in empleados.items():
        promedio=round(sum(puntajes)/len(puntajes),2)
        desempenho="Excelente"
        valor=round((100-minimo)/3,0)
        if promedio<minimo:
            desempenho="Deficiente"        
        elif promedio<(minimo+valor):
            desempenho="Regular"
        elif promedio<(minimo+valor*2):
            desempenho="Bueno" 
        else:
            desempenho="Excelente"
        
        tupla=(empleado,promedio,desempenho)
        if promedio>=minimo:
            lista_empleados.append(tupla)
        print(f"el empleado {empleado} tiene este promedio {promedio}")

        ordenado=sorted(lista_empleados,key=lambda x: (x[1],x[0]))

    return ordenado
```

```python
# Ejercicio 8

#comprimido = comprimir_texto(texto) 
#original = descomprimir_texto(comprimido) 

def comprimir_texto(texto:str)->str:
    resultado=[]
    caracter_actual=texto[0]
    contador=1

    for caracter in texto[1:]:
        if caracter == caracter_actual:
            contador+=1
        else:
            resultado.append(f"{caracter_actual}{contador}")
            contador=1
            caracter_actual=caracter
    
    resultado.append(f"{caracter_actual}{contador}")

    resultado="".join(resultado)
    return resultado



texto = 'aaabbccccdaa' 
print(comprimir_texto(texto))
```

```python
def descomprimir_texto(texto:str)->str:
    letras=texto[0::2]##para agrupar las posiciones pares
    numeros=texto[1::2]##para agrupar las posiciones impares
    resultado=[]
    for letra,numero in zip(letras,numeros):
        resultado.append(letra*int(numero))
        
    resultado="".join(resultado)
    return resultado
texto='a3b2c4d1a2'
print(descomprimir_texto(texto))
```

```python
# Ejercicio 9

def transferir(cuentas:dict[str:dict[str:str|int]],origen:str,destino:str,monto:int)->dict[str:str|int]:
    
    if origen not in cuentas:
        return (f"Error, no hay cuenta origen: {origen}")
    if destino not in cuentas:
        return (f"Error, no hay cuenta destino: {destino}")
    if origen==destino:
        return("Error, las cuentas son iguales")
    if not isinstance(monto,int) or monto<0:
        return(f"El monto {monto} no es numerico o positivo")
    valor=cuentas[origen]['saldo']
    if monto>valor:
        return("No hay saldo suficiente")
    cuentas[origen]['saldo']-=monto
    cuentas[destino]['saldo']+=monto
    comprobante={
        "Estado":"Operacion exitosa",
        "Origen":cuentas[origen]['titular'],
        "Destino":cuentas[origen]['titular'],
        "Monto":monto,
        "Saldo restante":cuentas[origen]['saldo']
    }
    return comprobante
```

```python
cuentas = { 
'001': {'titular': 'Ana', 'saldo': 1500}, 
'002': {'titular': 'Luis', 'saldo': 800} 
} 
transferir(cuentas,'001','002', 100)
```

```python
# Ejercicio 10

def reserva_asientos(sala:list[list[str]],asiento_solicitado:list[tuple[int]])->str:
    sala_fila=len(sala)
    sala_columnas=[]
    for fila in sala:
        sala_columnas.append(len(fila))
    
    for fila,columna in asiento_solicitado:
        if sala_fila<fila:
            return("No existe esa fila")
        if max(sala_columnas)<columna:
            return("No existe esa columna")
        asiento=sala[fila-1][columna-1]
        if asiento=='O':
            return(f"El asiento en la fila {fila} y columna {columna} esta ocupado ")
        
    for fila,columna in asiento_solicitado:    
        sala[fila-1][columna-1]='O'

    reserva={
        "Estado":"Asientos reservados",
        "Asientos resevador":asiento_solicitado,
        "Sala actualizada":sala
    }
    return reserva
```

```python
sala = [['L', 'L', 'O', 'L'], 
['O', 'L', 'L', 'L'], 
['L', 'O', 'L', 'L']] 
asientos = [(1, 2), (2, 3), (3, 1)] 
print(sala)
reserva_asientos(sala,asientos)
```
