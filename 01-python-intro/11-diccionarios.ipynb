```python
##ejemplo de como hacer diccionarios anidados
paises={}
paises["peru"]={}

paises["peru"]["capital"]="lima"
paises["peru"]["moneda"]="pen"
paises["peru"]["presidenta"]="keiko"

paises["chile"]={
    "capital":"santiago",
    "moneda":"peso chileno",
    "presidenta":"kazt"
}
print(paises)
```
